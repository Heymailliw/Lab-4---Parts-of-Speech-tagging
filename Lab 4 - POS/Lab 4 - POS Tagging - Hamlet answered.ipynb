#### Print the List of stopwords from NLTK in English <br> assign them to a variable called: stopwords
hint:  
stopwords = stopwords.words("english")
then print them.

```python
text_file = open("hamlet.txt")


text = text_file.read()


words = word_tokenize(text)


words_no_punc = []


for w in words:
    if w.isalpha():  
        words_no_punc.append(w.lower())


print("Words without punctuation:")
print(words_no_punc)


print("Length of words without punctuation:", len(words_no_punc))


stopwords = nltk.corpus.stopwords.words("english")
print("List of Stopwords in English:")
print(stopwords)
```
